```python
json_path = "/data/shenyuge/lingyue-data-process/logic_chain/gpt_result.json"
output_path = "/data/shenyuge/lingyue-data-process/logic_chain/merge_result.json"
# produce_quadruples(json_path, output_path)
category_res = category_lengths(output_path)
print(category_res)
```

```python
%matplotlib inline
```

```python
fig, axis = plt.subplots(2,3, figsize=(10,6),sharey=True)
axis = axis.flatten()
for i, category in enumerate(category_res.keys()):
        sns.boxplot(category_res[category], ax=axis[i])
        axis[i].set(ylabel="Count", xlabel=category)
fig.suptitle("Boxplot for Logic Chain Lengths")
fig.tight_layout()
```

```python
fig, axis = plt.subplots(2,3, figsize=(10,6))
axis = axis.flatten()
for i, category in enumerate(category_res.keys()):
        mean_value = category_res[category].mean()
        median_value = np.median(category_res[category])
        sns.histplot(category_res[category], ax=axis[i], binwidth=1)
        axis[i].set(ylabel="Count", xlabel=category)
        axis[i].axvline(mean_value, color='k', lw=2, label=f'Mean: {mean_value:.2f}')
        axis[i].axvline(median_value, color='k', ls='--', lw=2, label=f'Median: {median_value:.2f}')
        axis[i].legend()
fig.suptitle("Histgram for Logic Chain Lengths")
fig.tight_layout()
```

```python
# Add vertical lines for mean and median
ax.axvline(mean_value, color='red', linestyle='--', label=f'Mean: {mean_value:.2f}')
ax.axvline(median_value, color='green', linestyle='-.', label=f'Median: {median_value:.2f}')

# Add legend
ax.legend()

# Add labels and title
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Histogram with Mean and Median Lines')

# Show plot
plt.show()
```
